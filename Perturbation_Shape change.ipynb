```
# Resize the original image

from PIL import Image
# The path is changed for every fruit we like apply a shape change

path = "/content/gdrive/My Drive/Colab Notebooks/fruits/Test/Lemon/"

dirs = os.listdir( path )

def resize():
    for item in dirs:
        if os.path.isfile(path+item):
            im = Image.open(path+item)
            f, e = os.path.splitext(path+item)
            imResize = im.resize((100,80), Image.ANTIALIAS)
            imResize.save(f + ' resized_lime.jpg', 'JPEG', quality=90)

resize()
```

```
# then we create a folder for the modified images 1. for the resized images 2. for resized image concatenated with white space

def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
    except OSError:
        print ('Error: Creating directory. ' +  directory)
        

# Example


createFolder('./My Drive/Colab Notebooks/fruits/Test_Shape/Lemon/shape_lemon')
createFolder('./My Drive/Colab Notebooks/fruits/Test_Shape/Lemon/concat_resized_lemon')

```

```
# We then move the resized imaged to a different folder by searching the folder specifically by the name

import shutil
import os
import fnmatch

source = "/content/gdrive/My Drive/Colab Notebooks/fruits/Test/Lemon/"

destination_shape_folder = "/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/Lemon/shape_lemon/"

for files in os.listdir(source):
  if fnmatch.fnmatch(files, '*resized_lime.jpg'):
    shutil.move( source + files, destination_shape_folder)
```

```
# We then create a white space of the required size

from PIL import Image, ImageDraw

img = Image.new('RGB', (60, 30), color = 'white')
img = img.resize((100,100),Image.ANTIALIAS)
img.save("/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/white_space.jpg", "JPEG", quality=75)

rect = Image.open("/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/white_space.jpg")
rect2 = rect.resize((100,10), Image.ANTIALIAS)
rect2.save("/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/rect_10_r.jpg", "JPEG", quality=75)
rect2.save("/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/rect_10_l.jpg", "JPEG", quality=75)

```

```
# The final step was to concatenate the resized image and the white space

import cv2 

from PIL import Image, ImageDraw
from matplotlib import pyplot as plt

white_image_2 = cv2.imread('/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/rect_10_r.jpg')
white_image_3 = cv2.imread('/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/rect_10_l.jpg')



path = "/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/Lemon/shape_lemon/"
final_path = "/content/gdrive/My Drive/Colab Notebooks/fruits/Test_Shape/Lemon/concat_resized_lemon/"
dirs = os.listdir( path )

def concat_image():
    for item in dirs:
        if os.path.isfile(path+item):
            
            main_image = cv2.imread(path+item)
            
            f, e = os.path.splitext(path+item)
            
            filename = ('concat_img_'+ item)

            # The function cv2.hconcat (horizontally concatenates the images) and 
            # cv2.vconcat (vertically concatenates images) are used  interchangibly depending 
            # on if we want to stretch an image or collapse it 
                        
                       
            concat_img = cv2.vconcat([white_image_3, main_image , white_image_2])
            #concat_img = cv2.hconcat([white_image_3, main_image , white_image_2])
            
            
                                                              
            cv2.imwrite(final_path + filename , concat_img)
        
concat_image()
```
